```python
#This notebook is used for Dynamical Systems MTH 487
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import dslib as ds
```

```python
#Creating a slope field for xdot = 1-4x(1-x)

plt.figure(figsize=(8,6))
def f(x, t): #Define a function f(x(t))
    return 1-4*x*(1-x)

#Setting axis properties
plt.axvline(0, color='black', lw=1)
plt.axhline(0, color='black', lw=1)

#Slope field (function, horizontal bounds, vertical bounds, field line colors)
ds.plot_slope_field(f, hbounds=(-1, 5, 21), vbounds=(-1, 3, 21), color='darkblue')
plt.xlabel('$t$')
plt.ylabel('$x$')
plt.title('Slope field for the differential equation $\\dot{x}=1-4x(1-x)$')
None
```

```python
#Creating a slope field for xdot = 1-x^2

plt.figure(figsize=(8,6))
def f(x, t): #Define a function f(x(t))
    return 1 - x**2

#Setting axis properties
plt.axvline(0, color='black', lw=1)
plt.axhline(0, color='black', lw=1)

#Solution plot (function, time for initial condition, list of x values for initial conditions, x-axis limits, field line colors)
ds.plot_slope_field(f, hbounds=(-1, 5, 21), vbounds=(-3, 3, 21), color='darkblue')
plt.xlabel('$t$')
plt.ylabel('$x$')
plt.title('Slope field for the differential equation $\\dot{x}=1-x^2$')
None
```

```python
#Creating a solution plot for xdot = 1-x^2

plt.figure(figsize=(8,6))
def f(x, t): #Define a function f(x(t))
    return x-x**3

#Setting axis properties
plt.axvline(0, color='black', lw=1)
plt.axhline(0, color='black', lw=1)

#Solution plot (function, time for initial condition, list of x values for initial conditions, x-axis limits, field line colors)
ds.plot_solutions_1d(f, t0=0, inits=[0.5, -0.5], tmin=-5, tmax=5, color='royalblue', lw=2,
                    fixed_points=[-1, 0, 1])
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.title(f'Solutions of the DE $\\dot{{x}}=x-x^3$')
None
```

```python
#Creating a slope field for xdot = x(2-x)

plt.figure(figsize=(8,6))
def f(x, t): #Define a function f(x(t))
    return 1 - x - np.exp(-x)

#Setting axis properties
plt.axvline(0, color='black', lw=1)
plt.axhline(0, color='black', lw=1)

#Slope field (function, horizontal bounds, vertical bounds, field line colors)
ds.plot_slope_field(f, hbounds=(-1, 5, 21), vbounds=(-1, 3, 21), color='darkblue')
plt.xlabel('$r$')
plt.ylabel('$x$')
plt.title('Slope field for the differential equation $\\dot{x}=x(2-x)$')
None
```
